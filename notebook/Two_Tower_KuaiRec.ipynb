```python
print(">> example of Two-Tower dataset")
two_tower_dataset[0]
```

```python
user_vocab_size = {col: train_data[col].nunique() for col in user_cate_cols}
item_vocab_size = {col: train_data[col].nunique() for col in item_cate_cols}

embedding_dim = 128

user_first_in_features = embedding_dim * len(user_cate_cols) + len(user_num_cols)
user_layer_dims = [(user_first_in_features, 128), (128, 64)]
item_first_in_features = embedding_dim * len(item_cate_cols) + len(item_num_cols)
item_layer_dims = [(item_first_in_features, 128), (128, 64)]


user_tower_model = TowerModel(
    embedding_vocab_size=user_vocab_size,
    embedding_dim=embedding_dim,
    num_continuous_feature=len(user_num_cols),
    layer_dims=user_layer_dims
)

item_tower_model = TowerModel(
    embedding_vocab_size=item_vocab_size,
    embedding_dim=embedding_dim,
    num_continuous_feature=len(item_num_cols),
    layer_dims=item_layer_dims
)
```

```python
two_tower = TwoTower(user_tower_model, item_tower_model)
```

```python
import argparse
import torch
```

```python
def define_argparser():
    parser = argparse.ArgumentParser()

    parser.add_argument("--project", default="Two-Tower")
    parser.add_argument(
        "--batch_size",
        type=int,
        default=256,
        help="input batch size for training (default: 256)",
    )
    parser.add_argument(
        "--epochs", type=int, default=10, help="number of epochs to train (default: 10)"
    )
    parser.add_argument("--cuda", type=int, default=0, help="0 for cpu -1 for all gpu")
    config = parser.parse_args(args=[]) # add [] in jupyter notebook
    if config.cuda == 0 or torch.cuda.is_available() is False:
        config.cuda = 0

    return config

config = define_argparser()

config
```

```python
lit_two_tower = LitTwoTower(two_tower, config)
```

```python
from torch.utils.data import random_split, DataLoader
```

```python
train_ratio = 0.8

seed = torch.Generator().manual_seed(42)
train_dataset, valid_dataset = random_split(
    two_tower_dataset, [train_ratio, 1.0 - train_ratio], generator=seed
)
```

```python
train_loader = DataLoader(train_dataset, batch_size=config.batch_size)
valid_loader = DataLoader(valid_dataset, batch_size=config.batch_size)
```

```python
import pytorch_lightning as pl
```

```python
early_stopping_callback = pl.callbacks.EarlyStopping(
    monitor="validation/loss", mode="min", patience=3
)

trainer = pl.Trainer(
    log_every_n_steps=10,
    max_epochs=config.epochs,
    deterministic=True,
    callbacks=[early_stopping_callback],
    accelerator='cpu'
)
```

```python
trainer.fit(model=lit_two_tower, train_dataloaders=train_loader, val_dataloaders=valid_loader)
```
